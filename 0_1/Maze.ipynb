```python
def main():
    import copy
        
    layer = " "
    maze_s = []
    i = 0
    
    print("Нажмите enter для пустой строчки чтобы закончить ввод лабиринта")
    while layer != "":
        i += 1
        layer = str(input("Введите " + str(i) + " строку лабиринта"))
        maze_s.append(layer)
    
    '''
    maze_s = ["########",
            "# # ####",
            "# #N#   ",
            "# # # ##",
            "# # # ##",
            "#      #",
            "########"]
    '''
    
    for i in range(len(maze_s)):
        if maze_s[i].find("N") != -1:
            Noora = [maze_s[i].find("N"), i]
    
    maze_s[Noora[1]] = maze_s[Noora[1]][:Noora[0]] + " " + maze_s[Noora[1]][Noora[0] + 1:]
    for e in maze_s:
        print(e)
    
    exit = False
    
    def way_out(Noora_pos: list, maze: list):
        global exit
        if exit == True:
            return True
        if Noora_pos[0] == len(maze[Noora_pos[1]]) - 1 or Noora_pos[0] == 0 or Noora_pos[1] == len(maze) - 1 or Noora_pos[1] == 0:
            exit = True
    
        n_maze = copy.deepcopy(maze)
        if Noora_pos[0] < len(maze[Noora_pos[1]]) - 1:
            if maze[Noora_pos[1]][Noora_pos[0] + 1] == " ":
                n_maze[Noora_pos[1]] = maze_s[Noora_pos[1]][:Noora_pos[0]] + "#" + maze_s[Noora_pos[1]][Noora_pos[0] + 1:]
                way_out([Noora_pos[0] + 1, Noora_pos[1]], n_maze)
        if Noora_pos[0] > 0:
            if maze[Noora_pos[1]][Noora_pos[0] - 1] == " ":
                n_maze[Noora_pos[1]] = maze_s[Noora_pos[1]][:Noora_pos[0]] + "#" + maze_s[Noora_pos[1]][Noora_pos[0] + 1:]
                way_out([Noora_pos[0] - 1, Noora_pos[1]], n_maze)
        if Noora_pos[1] < len(maze) - 1:
            if maze[Noora_pos[1] + 1][Noora_pos[0]] == " ":
                n_maze[Noora_pos[1]] = maze_s[Noora_pos[1]][:Noora_pos[0]] + "#" + maze_s[Noora_pos[1]][Noora_pos[0] + 1:]
                way_out([Noora_pos[0], Noora_pos[1] + 1], n_maze)
        if Noora_pos[1] > 0:
            if maze[Noora_pos[1] - 1][Noora_pos[0]] == " ":
                n_maze[Noora_pos[1]] = maze_s[Noora_pos[1]][:Noora_pos[0]] + "#" + maze_s[Noora_pos[1]][Noora_pos[0] + 1:]
                way_out([Noora_pos[0], Noora_pos[1] - 1], n_maze)
        return exit
    
    print(way_out(Noora, maze_s))

if __name__ == "__main__":
    main()
```
